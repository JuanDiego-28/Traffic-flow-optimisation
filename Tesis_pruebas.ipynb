```python
import osmnx as ox
import networkx as nx
import matplotlib.pyplot as plt

# Retrieve a street network for the specified location
G = ox.graph_from_point((-0.18303, -78.48465), dist=4000, network_type='drive')

#Convert to dataframe
data = nx.to_pandas_edgelist(G)

# Create a color mapping based on osmid
osmid_color_mapping = {
    24212067	: 'red',
    24783925: 'blue',
    24381370	: 'green'

    # Add more osmid-color mappings
}

# Extract node colors based on osmid
node_colors = [osmid_color_mapping[n] if n in osmid_color_mapping else 'gray' for n in G.nodes("osmid")]

# Draw the graph with node colors
ox.plot_graph(ox.project_graph(G), node_color=node_colors)
plt.show()
```
